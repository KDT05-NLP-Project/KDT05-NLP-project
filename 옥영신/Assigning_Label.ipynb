```python
import pandas as pd
import numpy as np
import torch

# 1부 대본 데이터 불러오기
script1 = pd.read_excel('../data/야인시대 1부 (1~50) 통합대본.xlsx')
script1 = script1.dropna()
script1.head()
# 2부 대본 데이터 불러오기
script2 = pd.read_excel('../data/야인시대 2부(51~124) 통합대본.xlsx')
script2 = script2.dropna()
script2.head()
# 데이터 합치기
script = pd.concat([script1, script2])
script
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(script['대사'])

# K-means 클러스터링
k = 7  # 클러스터의 수. 민족주의, 공산주의, 자유주의, 반공주의, 민주주의, 국가주의, unknown.
model = KMeans(n_clusters=k, random_state=42)
model.fit(X)

# 클러스터 레이블 예측
labels = model.predict(X)

# 결과를 데이터프레임에 추가
script['클러스터'] = labels
```

```python
# 클러스터별로 대사 출력
for i in range(k):
    print(f'클러스터 {i}번 대사')
    print(script.loc[script['클러스터'] == i, '대사'].sample(5))
    print()
```

```python
'''
KMeans는 내가 원하는 방식으로 클러스터가 나뉘지 않는다.
'''
```

```python
# 클러스터 컬럼 삭제
script = script.drop('클러스터', axis=1)
script
```

```python
# 구두점 제거
import string
script['대사'] = script['대사'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
script
```

```python
# 형태소로 나누기
from konlpy.tag import Komoran

komoran = Komoran()
script['형태소'] = script['대사'].apply(lambda x: komoran.morphs(x))
script
```

```python
# 조사 제거
import re
def remove_postpositions(morphemes):
    cleaned_morphemes = [morpheme for morpheme in morphemes if not re.match('.*(이|가|은|는|을|를|과|와|아|야|이여|여|라|이라|라고|이라고|에|에서|부터|까지|에게|로|으로|의|처럼|만|마저|조차|한테|로서|으로서|인|인가|인지).*', morpheme)]
    return cleaned_morphemes
```

```python
script['형태소'] = script['형태소'].apply(remove_postpositions)
script
```

```python
# 불용어 제거
from soynlp.normalizer import emoticon_normalize

def remove_unnecessary_sounds(morphemes):
    # 이모티콘 정규화 and 웃음 소리나 의미없는 소리 제거
    cleaned_morphemes = [emoticon_normalize(morpheme, num_repeats=2) for morpheme in morphemes if morpheme not in ['하하', '하하하', '흠', '음', '허허', '음하하', '흐흐흐', '흐', '흐흐', '으응', '허허허', '으', '으허허', '으허','음허허']]
    return cleaned_morphemes

script['형태소'] = script['형태소'].apply(remove_unnecessary_sounds)
```

```python
# 대사 이념 분류 함수
def classify_dialogue(dialogue):
    counts = {key: 0 for key in word_lists.keys()}
    for word in dialogue.split():
        for key, word_list in word_lists.items():
            if word in word_list:
                counts[key] += 1
    return max(counts, key=counts.get)

# 대사 분류
script['대사 레이블'] = script['대사'].apply(classify_dialogue)
```

```python
# 형태소 이념 분류 함수
def classify_morphemes(morphemes):
    counts = {key: 0 for key in word_lists.keys()}
    for morpheme in morphemes:  # iterate over the list directly
        for key, word_list in word_lists.items():
            if morpheme in word_list:
                counts[key] += 1
    return max(counts, key=counts.get)

# 형태소 분류
script['형태소 레이블'] = script['형태소'].apply(classify_morphemes)
```
